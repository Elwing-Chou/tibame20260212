````python
def is_leap_year(year):
    # TAB/shift+TAB
    # year = 2000
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False

is_leap_year(2000)
````
