````python
def is_leap_year(year):
    # TAB/shift+TAB
    # year = 2000
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False
````

````python
def is_date_valid(d):
    # d = 20240229
    # 243 % 100 = 43
    # 243 // 100 = 2
    day = d % 100
    remain = d // 100
    month = remain % 100
    year = remain // 100
    # print(year, month, day)

    # 年檢定/月檢定
    # 十二個月該怎麼剪定日
    if year >= 1:
        if 1 <= month <= 12:
            days_per_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
            day_max = days_per_month[month-1]
            # if month == 2 and is_leap_year(year) == True
            if month == 2 and is_leap_year(year):
                day_max = day_max + 1
            if 1 <= day <= day_max:
                return True
    return False

# is_date_valid(20240230)
````

````python
def count_fully_squared_2(start, maxcount):
    # 20241001 = 4499 ** 2
    count = 0
    # start = 4499

    result = []
    while True:
        d = start ** 2
        if is_date_valid(d):
            # print(d)
            result.append(d)
            count = count + 1
        if count == maxcount:
            break
        start = start + 1
    return result

count_fully_squared_2(4499, 20)
````

````python
# 確認開根號是否為整數
# 只要兩個小數差不多(10**-6=0.000001), 他們就是一樣的數字
# 1. 把數字開根號 2. 把開根號的數字直接取四捨五入
# case1. 3.0000001  3 -> 0.0000001 我認為開根號完就是3
# case2. 3.234      3 -> 0.234
# 1和2相減 取絕對值
def is_sqrt_int(n):
    # n = 104976
    n_sqrt = n ** 0.5
    n_sqrt_round = round(n_sqrt)
    if abs(n_sqrt-n_sqrt_round) <= 10 ** -6:
        return True
    else:
        return False
is_sqrt_int(890)
````
