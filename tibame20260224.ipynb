```python
def is_leap_year(year):
    # TAB/shift+TAB
    # year = 2000
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False
```

```python
def is_date_valid(d):
    # d = 20240229
    # 243 % 100 = 43
    # 243 // 100 = 2
    day = d % 100
    remain = d // 100
    month = remain % 100
    year = remain // 100
    # print(year, month, day)

    # 年檢定/月檢定
    # 十二個月該怎麼剪定日
    if year >= 1:
        if 1 <= month <= 12:
            days_per_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
            day_max = days_per_month[month-1]
            # if month == 2 and is_leap_year(year) == True
            if month == 2 and is_leap_year(year):
                day_max = day_max + 1
            if 1 <= day <= day_max:
                return True
    return False

# is_date_valid(20240230)
```
