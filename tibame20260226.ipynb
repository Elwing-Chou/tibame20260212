```python
# 想法2
def is_leap_year(year):
    # TAB/shift+TAB
    # year = 2000
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False

def is_date_valid(d):
    # d = 20240229
    # 243 % 100 = 43
    # 243 // 100 = 2
    day = d % 100
    remain = d // 100
    month = remain % 100
    year = remain // 100
    # print(year, month, day)

    # 年檢定/月檢定
    # 十二個月該怎麼剪定日
    if year >= 1:
        if 1 <= month <= 12:
            days_per_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
            day_max = days_per_month[month-1]
            # if month == 2 and is_leap_year(year) == True
            if month == 2 and is_leap_year(year):
                day_max = day_max + 1
            if 1 <= day <= day_max:
                return True
    return False

def count_fully_squared_2(start=4499, maxcount=10):
    # 20241001 = 4499 ** 2
    count = 0
    # start = 4499

    result = []
    while True:
        d = start ** 2
        if is_date_valid(d):
            # print(d)
            result.append(d)
            count = count + 1
        if count == maxcount:
            break
        start = start + 1
    return result
```

```python
# 想法1
def is_sqrt_int(n):
    # n = 104976
    n_sqrt = n ** 0.5
    n_sqrt_round = round(n_sqrt)
    if abs(n_sqrt-n_sqrt_round) <= 10 ** -6:
        return True
    else:
        return False

import datetime as dt
# 開始日期
start = dt.datetime(2024, 10, 1)
delta = dt.timedelta(days=1)
# 往後迴圈
count = 0
while True:
    # 日期->字串(20241001)->整數
    n = int(start.strftime("%Y%m%d"))
    # 你要確認整數開根號是否是整數(下面說)
    if is_sqrt_int(n):
        # 如果是整數就是個完全平方日
        print(n)
        count = count + 1
    if count == 10:
        break
    start = start + delta
```
